# 오즈와 오즈비 (Odds & Odds Ratio)


## 01.오즈와 오즈비 개념 이론
- 배우는 이유

    - 로지스틱 회귀는 예측 결과뿐 아니라 feature가 결과를 어느 방향으로 움직이는지도 해석할 수 있음.
    - 오즈비는 로지스틱 회귀 계수를 몇 배 변화 관점으로 바꿔 설명하는 도구임.
    - 정확도만으로는 왜 이런 예측이 나왔는지 설명하기 어려워 모델 해석을 위해 배움.

- 어디에 사용되는가?

    - 의료/헬스케어: 흡연, 약 복용, 검사 수치가 질병 발생 오즈에 어떤 영향을 주는지 해석.
    - 마케팅/서비스: 연령, 행동 로그, 방문 패턴이 구매/이탈 오즈에 미치는 영향 해석.
    - 금융/리스크: 연체, 부도, 사기 발생 가능성을 설명 가능한 방식으로 분석.
### 01-01. 오즈 (Odds)
- 오즈는 어떤 사건이 발생할 확률과 발생하지 않을 확률의 비율이다.
- 수식: `오즈 = P(사건 발생) / P(사건 미발생)`
- 예: 질병에 걸릴 확률이 0.8이라면 오즈는 0.8 / 0.2 = 4

### 01-02. 오즈비 (Odds Ratio)
- 두 집단 간 오즈의 비율을 의미하며, 사건이 한 집단에서 다른 집단보다 얼마나 자주 발생하는지를 비교한다.
- 수식: `오즈비 = (a / b) / (c / d) = (a*d) / (b*c)`
  - a: 실험군에서 사건 발생 수
  - b: 실험군에서 사건 미발생 수
  - c: 대조군에서 사건 발생 수
  - d: 대조군에서 사건 미발생 수

### 01-03.해석 기준
| OR 값 | 의미 |
|--------|------|
| OR = 1 | 두 그룹의 사건 발생 가능성이 동일함 |
| OR > 1 | 앞 그룹에서 사건 발생 가능성이 더 높음 |
| OR < 1 | 뒤 그룹에서 사건 발생 가능성이 더 높음 |



## 02.약복용여부에 따른 발병여부

### 02-01.예시 상황
| 그룹       | 질병 발생 | 질병 없음 | 전체 |
|------------|------------|------------|-------|
| 약 복용    | 20         | 80         | 100   |
| 비복용     | 40         | 60         | 100   |

- 약 복용 그룹의 오즈: 20 / 80 = 0.25
- 비복용 그룹의 오즈: 40 / 60 = 0.667
- 오즈비: (20 * 60) / (80 * 40) = 1200 / 3200 = 0.375

> 해석: 약 복용 그룹은 비복용 그룹보다 질병 발생 오즈가 더 낮음 → 약이 효과 있음


```python
# statsmodels 설치
%pip install statsmodels
```

```python
from statsmodels.stats.contingency_tables import Table2x2

# 작성: 약 복용/비복용 그룹의 질병 발생 여부를 2x2 교차표로 작성하세요.
# 힌트: [[약 복용 그룹의 질병 발생 수, 약 복용 그룹의 질병 없음 수],
#       [비복용 그룹의 질병 발생 수, 비복용 그룹의 질병 없음 수]]
table = [[20, 80],
         [40, 60]]

# 작성: Table2x2()로 교차표 분석 객체를 생성하세요.
result = Table2x2(table)

# 작성: 오즈비와 95% 신뢰구간을 출력하세요.
print(f'odds ratio: {result.oddsratio:.3f}')

# oddsratio_confint() : 오즈비의 95% 신뢰구간(범위)
# - ci_low ~ ci_high 사이에 95% 확률로 오즈비가 존재한다는 뜻 XXXXXXXX
# - 같은 방식으로 표본을 여러번 뽑아 신뢰구간을 만들면
#   그 구간들 중 95%는 진짜 오즈비가 포함된다는 뜻
ci_low, ci_high = result.oddsratio_confint()
print(f'95% 신뢰구간: {ci_low:.3f} ~ {ci_high:.3f}')
```

## 03. 로지스틱 회귀에서의 오즈비 해석

### 03-01. 예제: 흡연 여부에 따른 질병 발생

**회귀 계수 (coef)**는 로그 오즈비 (log odds ratio) -> **np.exp(coef)**는 오즈비

* 오즈비 > 1 → 양의 상관관계 (흡연하면 질병 확률 증가)
* 오즈비 < 1 → 음의 상관관계 (흡연하면 질병 확률 감소)
  
> 해석: np.exp(coef)가 오즈비. 예: exp(1.2) = 3.32 → 흡연자는 질병 오즈(질병 발생할 비율)가 3.3배 


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

# 작성: smoker, disease 컬럼을 가진 예제 DataFrame을 생성하세요.
# smoker: 흡연 여부, disease: 질병 발생 여부
# smoker: 흡연 여부. 1은 흡연, 0은 비흡연을 의미함.
# disease: 질병 발생 여부. 1은 발생, 0은 미발생을 의미함.
df = pd.DataFrame({
    'smoker': [1, 1, 1, 1, 0, 0, 0, 0],
    'disease': [1, 1, 0, 1, 0, 0, 0, 1]
})

# X는 입력 feature이므로 DataFrame 형태로 유지함.
# y는 예측해야 하는 정답이므로 Series 형태로 사용함.
X = df[['smoker']]
y = df['disease']

# 작성: feature X와 target y를 분리하세요.


# 작성: LogisticRegression 모델을 생성하고 학습시키세요.
log_reg =  LogisticRegression()
log_reg.fit(X, y)

# 작성: 학습된 회귀계수를 출력하세요.
print("회귀계수:", log_reg.coef_)

# 작성: 회귀계수를 오즈비로 변환해 출력하세요.
# 힌트: np.exp(coef)
print("오즈비:", np.exp(log_reg.coef_))

# 오즈비가 1보다 큼 == 흡연자가 비흡연자보다 질병 발생 확률이 높다(1.95배)
```

## 04. 범주형 변수에서 오즈비

### 04-01. 교차표 기반 예시
| 성별 | 구매 | 비구매 |
|------|------|--------|
| 남자 | 30   | 70     |
| 여자 | 50   | 50     |

**오즈비가 0.429**
→ 남성은 여성에 비해 구매할 오즈가 더 낮다 (약 0.43배)

**95% 신뢰구간이 (0.229, 0.805)**
→ 이 구간에 1이 포함되지 않음(1이 포함되면, 오즈비가 모호하다는 것. 1보다 작거나, 커야한다)
→ 통계적으로 유의미하게 여성의 구매 오즈가 높다고 말할 수 있음

🧠 결론
- 여성은 남성보다 구매할 가능성이 높다.
- 오즈 기준으로 보면 약 2.3배 (1 / 0.429 ≈ 2.33) 더 많이 구매함
- 신뢰구간도 1보다 작으므로 유의미한 차이임


```python
from statsmodels.stats.contingency_tables import Table2x2

# 작성: 성별에 따른 구매/비구매 교차표를 작성하세요.
# 힌트: [[남자 구매 수, 남자 비구매 수],
#       [여자 구매 수, 여자 비구매 수]]
table = [[30, 70], [50, 50]]

# 작성: Table2x2()로 오즈비 분석 객체를 생성하세요.
result = Table2x2(table)

# 작성: 오즈비와 신뢰구간을 출력하세요.
print(f'odds ratio: {result.oddsratio:.3f}')
ci_low, ci_high = result.oddsratio_confint()
print(f'신뢰구간: {ci_low:.3f} ~ {ci_high:.3f}')
```

## 05. 다범주형 범주 변수의 오즈비

### 05-01. 예제: 직업별 구매 여부

> 해석:
> - engineer의 오즈비: teacher 대비 구매 가능성이 몇 배인지
> - student의 오즈비: teacher 대비 낮은지 높은지 판단


```python
# 직업처럼 범주가 3개 이상인 feature를 원핫 인코딩하고 기준 범주 대비 오즈비 확인.
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

# job: teacher, engineer, student 세 범주를 가진 feature임.
# buy: 구매 여부 target임.
df = pd.DataFrame({
    'job': ['teacher', 'engineer', 'student', 'teacher', 'student', 'engineer',
            'teacher', 'student', 'engineer', 'student'] * 10,
    'buy': [1, 1, 0, 1, 0, 1, 0, 1, 0, 0] * 10
})

# pd.get_dummies(): 문자 범주를 범주별 0/1 컬럼으로 변환함.
# astype(int): True/False로 만들어진 더미 컬럼을 1/0 정수로 바꿈.
df_encoded = pd.get_dummies(df, columns=['job']).astype(int)

# display(): Jupyter 노트북에서 DataFrame을 표 형태로 보기 좋게 출력함.
display(df_encoded)

# teacher를 기준 범주로 두기 위해 job_teacher 컬럼은 모델 입력에서 제외함.
# teacher인 행은 job_engineer=0, job_student=0으로 표현됨.
X = df_encoded[['job_engineer', 'job_student']]
y = df_encoded['buy']

model = LogisticRegression()
model.fit(X, y)

# coef_[0]에는 job_engineer, job_student 순서의 로그 오즈 계수가 들어 있음.
coef = model.coef_[0]

# exp(coef)를 계산하면 teacher 대비 각 직업의 구매 오즈비가 됨.
odds_ratios = np.exp(coef)

print("회귀계수:", coef)
print("오즈비:", odds_ratios)
```
